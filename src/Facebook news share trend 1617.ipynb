```python
import glob

files = glob.glob('./newsdata/newsreveal/*.json')
```

```python
import json
def read_json_file(path_to_file):
    with open(path_to_file) as p:
        rows=[json.loads(line) for line in p]
    return rows
```

```python
import pandas as pd

data=[]
for path in files:
    data.extend(read_json_file(path))    
    

df = pd.DataFrame(data)
```

```python
len(df)
```

```python
df.drop(['_id','author', 'naver_section'], axis = 1, inplace = True)
```

```python
df['year'] = df['date'].apply(lambda x: x[0] )
df['month'] = df['date'].apply(lambda x: x[1] )
df['datetime'] = df['date'].apply(lambda x: "-".join(x) )
df['total'] = df['shares'].apply(lambda x: x[0] )
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
media_total = df.groupby('media').org_url.count()
media_total.sort_values(ascending=False)
```

```python
media_total_by_year = df.groupby([pd.Grouper(key='datetime', freq='Y'),'media'])['org_url'].count()
media_total_by_year.unstack(level=-1)
```

```python
media_sum = df.groupby('media')['total'].sum()
media_sum.sort_values(ascending=False)
```

```python
media_sum_by_year = df.groupby([pd.Grouper(key='datetime', freq='Y'),'media'])['total'].sum()
media_sum_by_year.unstack(level=-1)
```

```python
media_mean = df.groupby('media')['total'].mean()
media_mean.sort_values(ascending=False)
```

```python
media_mean_by_year = df.groupby([pd.Grouper(key='datetime', freq='Y'),'media'])['total'].mean()
media_mean_by_year.unstack(level=-1)
```

```python
group_total=df.groupby([pd.Grouper(key='datetime', freq='M'), 'media'])['total'].sum()
unstacked_total=group_total.unstack(level=-1)
```

```python
unstacked_total.reset_index(level=0, inplace=True)
unstacked_total
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()
```

```python
p1 = figure(plot_width=750, plot_height=500,x_axis_type='datetime')
p1.grid.grid_line_alpha=0.3
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Count'
p1.line(unstacked_total['datetime'], unstacked_total['chosun'], color='#f90202', legend='',line_width=2)
p1.line(unstacked_total['datetime'], unstacked_total['joins'], color='#E2A00B', legend='',line_width=2)
p1.line(unstacked_total['datetime'], unstacked_total['donga'], color='#1cefef', legend='',line_width=2)
p1.line(unstacked_total['datetime'], unstacked_total['hani'], color='#01996e', legend='',line_width=2, line_dash='dashed')
p1.line(unstacked_total['datetime'], unstacked_total['kh'], color='#18407D', legend='',line_width=2, line_dash='dashed')
p1.line(unstacked_total['datetime'], unstacked_total['jtbc'], color='#F48179', legend='JTBC',line_width=2, line_dash='dotted')
p1.line(unstacked_total['datetime'], unstacked_total['sbs'], color='#1D3A6E', legend='SBS',line_width=2, line_dash='dotted')



show(p1)
```
