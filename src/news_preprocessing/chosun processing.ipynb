````python
import glob

files = glob.glob('./data/2017/*/chosun/*.json')
````

````python
import json
def read_json_file(path_to_file):
    with open(path_to_file) as p:
        rows=[json.loads(line) for line in p]
    return rows
````

````python
import pandas as pd

data=[]
for path in files:
    data.extend(read_json_file(path))

df = pd.DataFrame(data)
````

````python
df=df.dropna(subset=['section'])
df=df.reset_index(drop=True)
````

````python
import re
def remove_chosun_extras(article):
    article = re.sub(r'\/\/[a-zA-Z가-힣 \n]+|{}|_flash_removeCallback\(\)', '', article)
    article = re.sub(r'\/[AP ]+연합뉴스|\/[가-힣 ]+캡처|\/[조선dbDB ]+', '', article)
    article = re.sub(r'\[조선닷컴 바로가기\](.*)', '', article)
    article = re.sub(r'\[[가-힣=, ]+[기자인턴]\]|\[[가-힣=() ]+[a-z0-9.]+@chosun.com\]', '', article)
    article = article.replace('- Copyrights ⓒ 조선일보 & chosun.com, 무단 전재 및 재배포 금지 -', '')
    article = article.replace('\n', ' ')
    return article
````
