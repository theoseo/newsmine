```python
import glob

files = glob.glob('./data/2017/*/ohmynews/*.json')
```

```python
import json
def read_json_file(path_to_file):
    with open(path_to_file) as p:
        rows=[json.loads(line) for line in p]
    return rows
```

```python
import pandas as pd

data=[]
for path in files:
    data.extend(read_json_file(path))

df = pd.DataFrame(data)
```

```python
import re
def remove_extras(article):   
    article = article.replace('▶','\n▶').replace('\t', '').replace('\n', ' ')#.split('\n')
    article = re.sub(r'\/\/[a-zA-Z가-힣 \n]+|{}|_flash_removeCallback\(\)', '', article)
    article = re.sub(r'저작권자\(c\)(.*)', '', article.strip())
    article = re.sub(r'ⓒ [a-zA-Z가-힣= ]+(제공|캡처)|ⓒ [a-zA-Z가-힣=]+', '', article)

    article = re.sub(r'\[[가-힣(오마이뉴스)=:, ]+\]', '', article)

    article = re.sub(r'\【오마이뉴스는(.*)\】', '', article)
    article = re.sub(r'▶ 해당 기사는 모바일 앱 모이(.*)', '', article)


    article = article.replace('\n', ' ')
    return article
```
